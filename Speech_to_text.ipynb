```python
from transformers import pipeline
```

```python
p = pipeline("automatic-speech-recognition")
```

```python
!pip show gradio
```

```python
import gradio as gr
```

```python
def transcribe(audio):
    text = p(audio)["text"]
    return text
```

```python
gr.Interface(
    title = 'OpenAI Whisper ASR Gradio Web UI',
    fn=transcribe,
    inputs=[
        gr.Audio(type="filepath")
    ],
    outputs=[
        "textbox"
    ],
    live=True).launch()
```
